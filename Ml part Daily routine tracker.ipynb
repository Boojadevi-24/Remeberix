```python
import time
import datetime
import playsound
from IPython.display import display, Javascript

# Define recommended time slots
recommended_schedule = {
    "Sleep": "22:00",
    "Exercise": "06:30",
    "Breakfast": "07:30",
    "Work/Study (Morning)": "09:00",
    "Lunch": "13:00",
    "Relaxation/Meditation": "14:00",
    "Work/Study (Afternoon)": "15:00",
    "Dinner": "20:00",
    "Medicine Intake": ["09:00", "13:00", "20:30"]
}

# Function to show a pop-up alert
def show_alert(message):
    display(Javascript(f'alert("{message}")'))

# Function to check time and play an alarm
def check_time_and_alert():
    current_time = datetime.datetime.now().strftime("%H:%M")  # Get current time in HH:MM format

    for activity, time_slot in recommended_schedule.items():
        if isinstance(time_slot, list):  # If multiple times exist (like medicine)
            for t in time_slot:
                if current_time == t:
                    print(f" Reminder: It's time for {activity} ({t})!")
                    show_alert(f"It's time for {activity}!")
                    playsound.playsound("/usr/share/sounds/alsa/Front_Center.wav")  # Default Linux sound
        else:
            if current_time == time_slot:
                print(f" Reminder: It's time for {activity} ({time_slot})!")
                show_alert(f"It's time for {activity}!")
                playsound.playsound("/usr/share/sounds/alsa/Front_Center.wav")

# Run the function every minute to check for reminders
while True:
    check_time_and_alert()
    time.sleep(60)  # Wait for 1 minute before checking again
```
